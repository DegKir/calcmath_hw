## 2. Оценка количества итераций

Оценим количество итераций для того чтобы найти решение с точностью $\varepsilon=10^{-5}$

$||x^{(n)}-x^*||\leq \frac{q^n}{1-q}||x^{(1)}-x^{(0)}||$, где $q=max_i |\lambda_i(I-P^{-1}A)|$

Загоним Матрицу $A$ в sympy :

```python
A=Matrix(matrix)
```

### 2.1 Поиск матрицы перехода

```python
#Немного хардкодинга чтобы получить диагональную матрицу для метода Якоби
P=eye(100)
P=P*10
```

```python
S = eye(100) - P.inv()*A
```

```python
print(S)
```

### 2.2 Степенной метод поиска максимального по модулю собственного значения

Итерация происходит по следующим формулам ($r_0$ выбирается произвольно)

$r_{k+1}=Sr_k$

$\lambda_1^{(k)}=\frac{(r_{k+1},r_{k})}{(r_k,r_k)}$

$r_{k+1}=\frac{r_{k+1}}{||r_{k+1}||}$

```python
a=[1]
for i in range(0,99):
    a.append(1)
a
```

```python
r_start=Matrix(a)
r_old=r_start
lambda_new = 0;
delta=1
lambda_old=10000
while(delta>0.000001):
    r_new=S*r_old
    lambda_new=(r_new.T*r_old)[0]/(r_old.T*r_old)[0]
    print(lambda_new)
    lenght=r_new.norm()
    r_new=r_new/lenght
    delta= abs(lambda_new-lambda_old)
    print(delta)
    lambda_old=lambda_new
    r_old=r_new
```

```python
lambda_new
```

```python
q_lambda=abs(lambda_new)
```

Таким образом получено максимальное по модулю собственно значение матрицы

$||x^{(n)}-x^*||\leq \frac{q^n}{1-q}||x^{(1)}-x^{(0)}||$, где $q=max_i |\lambda_i(I-P^{-1}A)|$

```python
def check_accuracy(delta_zero, n,q=q_lambda,epsilon=1E-5):
    if (pow(q,n)*delta_zero/(1-q) > epsilon ):
        return true
    return false
```

### 3.1 Стартуем с нулевого вектора, ищем разницу между первыми двумя полученными векторами для дальнейшей оценки погрешности :
