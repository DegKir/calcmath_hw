## 1. Реализация матрицы и алгоритма

```python
variables=[]
a=[1,10,1,1/10]
matrix=[]
matrix_row=[]
for i in range(0,100):
    matrix_row.append(0)
matrix_row[0]=10
matrix_row[1]=1
matrix_row[2]=1/10
matrix_row[3]=0
matrix_row
```

```python
#Создаем нашу матрицу
matrix=[]
matrix.append(matrix_row)
matrix_row = matrix_row[-1:]+matrix_row[:-1]
matrix_row[0]=1
matrix.append(matrix_row)
for i in range (0,96):
    #сдвиг строки на один на один элемент
    matrix_row = matrix_row[-1:]+matrix_row[:-1]
    #добавляем в матрицу
    matrix.append(matrix_row)
    
#добавляем последние две строчки
matrix_row = matrix_row[-1:]+matrix_row[:-1]
matrix_row[0]=0
matrix.append(matrix_row)

matrix_row = matrix_row[-1:]+matrix_row[:-1]
matrix_row[0]=0
matrix.append(matrix_row)
```

```python
#Создаем правый столбец 
right=[]
for i in range(1,100):
    right.append(i)
right
```

Итерационный процесс происходит по формуле $x_m^{(k)}=\frac{1}{a_{mm}}(f_m-a_{m1}x_1^{(k-1)}-...-a_{mm-1}x_{m-1}^{k-1}-a_{mm+1}x_{m+1}^{(k-1)}-...-a_{mn}x_n^{(k-1)})$

```python
def get_xm(m):
    amm=matrix[m][m]
    fm=right[m]
    summ=fm
    for i in range(1,100):
        summ-=matrix[m][i]*variables[i]
    return (summ/amm)
```

```python
for i in range(0,100):
    variables.append(0)
len(variables)
```

```python
limit=10
new_variables=variables
while (limit >= 1):
    for i in range(0,99):
        new_variables[i]=get_xm(i)
    variables=new_variables
    limit-=1
```

## 2. Оценка количества итераций

Оценим количество итераций для того чтобы найти решение с точностью $\varepsilon=10^{-5}$

$||x^{(n)}-x^*||\leq \frac{q^n}{1-q}||x^{(1)}-x^{(0)}||$, где $q=max_i |\lambda_i(I-P^{-1}A)|$

Загоним Матрицу $A$ в sympy :

```python
A=Matrix(matrix)
```

### 2.1 Поиск матрицы перехода

```python
#Немного хардкодинга чтобы получить диагональную матрицу для метода Якоби
P=eye(100)
P=P*10
```

```python
S = eye(100) - P.inv()*A
```

```python
print(S)
```

### 2.2 Степенной метод поиска максимального по модулю собственного значения

Итерация происходит по следующим формулам ($r_0$ выбирается произвольно)

$r_{k+1}=Sr_k$

$\lambda_1^{(k)}=\frac{(r_{k+1},r_{k})}{(r_k,r_k)}$

$r_{k+1}=\frac{r_{k+1}}{||r_{k+1}||}$

```python
a=[1]
for i in range(0,99):
    a.append(1)
a
```

```python
r_start=Matrix(a)
r_old=r_start
lambda_new = 0;
delta=1
lambda_old=10000
while(delta>0.000001):
    r_new=S*r_old
    lambda_new=(r_new.T*r_old)[0]/(r_old.T*r_old)[0]
    print(lambda_new)
    lenght=r_new.norm()
    r_new=r_new/lenght
    delta= abs(lambda_new-lambda_old)
    print(delta)
    lambda_old=lambda_new
    r_old=r_new
```
